<font size=5>Metoda, kjer razpišemo integral, je približno dvakrat hitrejša.

```python
import time

data_length = [1000,5000,10000,50000]

cas1 = []
cas2 = []

for n in data_length:
    
    r.seed(0)
    
    gauss = []

    for _ in range(n):
        gauss.append(r.gauss(0,1))
    zac = time.time()
    obn1(gauss,choice="kl")
    cas1.append(time.time()-zac)
    
    zac = time.time()
    obn2(gauss,choice="kl")
    cas2.append(time.time()-zac)
    
import matplotlib.pyplot as plt
plt.plot(data_length, cas1, label="obn1")
plt.plot(data_length, cas2, label="obn2")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))
plt.plot(data_length, cas1, label="obn1")
plt.plot(data_length, cas2, label="obn2")
plt.legend()
plt.grid()
plt.show()
```

<font size=4>Torej: namesto $\int p(x)*ln(p(x)/q(x)) dx$ moramo računati $\int p(x)*ln(p(x)) dx - \int p(x)*ln(q(x)) dx$, kjer 1. člen izračunamo pred iteracijo.
