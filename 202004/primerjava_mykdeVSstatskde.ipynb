```python
from kernel_estimators import kde
from GMM import GMM
from scipy.stats import gaussian_kde as gkde

gmm = GMM([[0,1],[5,1]],100000)

help(kde)
```

```python
from time import time

zac = time()
pdf1 = gkde(gmm)
print("Metoda stats.gaussian_kde: {} sek".format(str(time()-zac)))


zac = time()
pdf2 = kde(gmm)
print("Metoda kde: {} sek".format(str(time()-zac)))
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3,8,1000)

plt.plot(x, pdf1(x),label="stats")
plt.plot(x,pdf2(x),label="kde")

plt.legend()

plt.show()
```

<font size=4>Metodi se ujemata. Sam izračun je nekoliko histrejši pri moji metodi, izris je nekoliko histrejši z vgrajeno metodo (tj. ko računamo pdf(x))

<font size=4>Kasneje se je ugotovilo, da je vgrajena metoda boljša, ko pride do nekih iteracij, saj je precej hitrejša.
