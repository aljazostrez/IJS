```python
import random as r
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt

def f(x): return 1/x

def integral_MC(n):
    # n = stevilo tock
    tocke = []
    for _ in range(n):
        tocke.append((r.random()+1,r.random()))
    tockeIN = 0
    tockeOUT = 0
    for i in range(n):
        xi = tocke[i][0]
        yi = tocke[i][1]
        if yi <= f(xi):
            tockeIN += 1
        else:
            tockeOUT += 1
    
    return tockeIN/len(tocke)
```

```python
x = [1000*i for i in range(1,1001)]
MC = []
I = integrate.quad(f,1,2)[0]

for i in x:
    MC.append(abs(integral_MC(i)-I)/I)
    print(i)

tocna = [I for i in x]

plt.plot(x,MC,label="MC")
plt.plot(x,tocna,label="tocna")

plt.legend()
plt.show()
```
