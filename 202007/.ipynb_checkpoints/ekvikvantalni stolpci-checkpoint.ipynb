```python
import numpy as np
from equiquantal_bins import eqBins
import matplotlib.pyplot as plt
from scipy import stats

data = np.random.normal(0,1,60000)
pdf = lambda x: stats.norm.pdf(x,0,1)

st_stolpcev = 20

plt.figure(figsize=(15,8))
plt.hist(data, bins=eqBins(data, st_stolpcev), density=1)

x = np.linspace(-5,5,1000)
plt.plot(x,pdf(x))

plt.show()
```

# Izračuni divergenc

Beta porazdelitev je taka, da zavzame vse splošne oblike, hkrati pa ima končni nosilec, zato testiramo na tej porazdelitvi.

```python
tocna1 = lambda x: stats.beta.pdf(x,0.5,0.5)
tocna2 = lambda x: stats.beta.pdf(x,1,1)
tocna3 = lambda x: stats.beta.pdf(x,2,2)
tocna4 = lambda x: stats.beta.pdf(x,0.5,2)
tocna5 = lambda x: stats.beta.pdf(x,1,2)
tocna6 = lambda x: stats.beta.pdf(x,2,3)

x = np.linspace(0.01,0.99,1000)

plt.figure(figsize=(15,10))

plt.subplot(1,2,1)
plt.title(r"$\alpha = \beta$")
plt.plot(x,tocna1(x),label="a=0.5,b=0.5")
plt.plot(x,tocna2(x),label="a=1,b=1")
plt.plot(x,tocna3(x),label="a=2,b=2")
plt.legend()

plt.subplot(1,2,2)
plt.title(r"$\alpha \neq \beta$")
plt.plot(x,tocna4(x),label="a=0.5,b=2")
plt.plot(x,tocna5(x),label="a=1,b=2")
plt.plot(x,tocna6(x),label="a=2,b=3")
plt.legend()

plt.show()
```

```python
def my_beta(alpha,beta, n=600):
    print("BETA: a={}, b={}\n\n".format(str(alpha),str(beta)))
    
    tocna = lambda x: stats.beta.pdf(x,alpha,beta)

    data = np.random.beta(alpha,beta,n)
    
    ## TEST
    x = np.linspace(0,1,1000)
    plt.figure(figsize=(15,15))
    
    plt.subplot(221)
    plt.title("hist n=10")
    plt.hist(data,bins=10,density=1)
    plt.plot(x,tocna(x),label="tocna")
    
    plt.subplot(222)
    plt.title("hist n=20")
    plt.hist(data,bins=20,density=1)
    plt.plot(x,tocna(x),label="tocna")
    
    plt.subplot(223)
    plt.title("hist adapt n=10")
    plt.hist(data,bins=eqBins(data,10),density=1)
    plt.plot(x,tocna(x),label="tocna")

    plt.subplot(224)
    plt.title("hist adapt n=20")
    plt.hist(data,bins=eqBins(data,20),density=1)
    plt.plot(x,tocna(x),label="tocna")
    
    
    print("tocna, gkde, histogrami")
    
    plt.show()
```

```python
my_beta(1,1)
```

<font size=4>Pričakovano: pri uniformni porazdelitvi ne bomo veliko pridobili z vpeljavo ekvikvantalnih stolpcev.

```python
my_beta(0.5,0.5)
```

<font size=4>Pri porazdelitvah s špicami smo videli, da so ekvikvantalni precej boljši.

```python
my_beta(2,2)
```

<font size=4>Pri zgornji porazdelitvi bi bilo najboljše uporabiti Gaussian kernel density estimation. Sicer nam pa ekvikvantalni ne pomagajo prav dosti.

```python
my_beta(0.5,2)
```

<font size=4>Spet so veliko boljši ekvikvantalni stolpci, ker imamo špico.

```python
my_beta(1,2)
```

<font size=4>Pri zgornji porazdelitvi nam ekvikvantalni stolpci ne pomagajo prav dosti.

```python
my_beta(2,3)
```

<font size=4>Pri zadnjem primeru bi nam spet bolj prav prišla predstavitev z Gaussian kernel density estimation.

<font size=4>**Zaključek**: vbistvu nikoli ni najbolj optimalna izbira z ekvidistantnimi stolpci, je pa najbolj kompaktna. Mogoče je najboljša izbira v primeru, da ne vemo popolnoma nič o porazdelitvi, ki jo obravnavamo. Sicer pa je pri porazdelitvah s špicami najboljša uporaba ekvikvantnih stolpcev (takrat ne moramo uporabiti Gaussian kernel density estimation, saj pademo ven iz nosilca). Pri porazdelitvah s trebuhi, ki nimajo omejenih nosilcev, pa je najboljša uporaba Gaussian kernel density estimation.
